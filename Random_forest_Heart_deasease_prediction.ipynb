```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 20)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(criterion="gini",
                             max_depth = 6,
                             min_samples_split = 5,
                            random_state=5)
```

```python
'''
y_train = y_train.astype(float)
X_train[:, :] = X_train[:, :].astype(float)
'''
```

```python
X_train
```

```python
clf.fit(X_train, y_train)
```

```python
fi = clf.feature_importances_[0:13]

fi1 = pd.DataFrame({'Value': fi, 'Percentage': fi*100})


# sort the DataFrame by the percentage
df_sorted = fi1.sort_values(by='Percentage', ascending=False)

# create the plot
sns.set(style='whitegrid')
ax = sns.barplot(x='Percentage', y='Value', data=df_sorted, color='y' )
ax.set(xlabel='Percentage', ylabel='Value')

# display values inside bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate('{:.2f}'.format(height), (x + width/2, y + height/2), ha='center', va='center')

# increase figure size
plt.gcf().set_size_inches(10, 6)

# show plot
plt.show()

```

```python
#Delete terget inside the column name
clm = df.columns

# create the index object
clm = pd.Index(['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target'])

# delete the 'target' attribute
clm = clm.delete(clm.get_loc('target'))

# print the new index
print(clm)
```

```python
fi
```

```python
# create a DataFrame with the values and the column names
df = pd.DataFrame({'Feature': clm, 'Importance': fi})

# sort the DataFrame by the importance values
df_sorted = df.sort_values(by='Importance', ascending=False)

# calculate the percentage and add a new column to the DataFrame
df_sorted['Percentage'] = df_sorted['Importance'] * 100 / df_sorted['Importance'].sum()

# create the plot using Seaborn
sns.set(style='whitegrid')
ax = sns.barplot(x='Percentage', y='Feature', data=df_sorted, color='green')
ax.set(xlabel='Percentage', ylabel='Feature')

# increase the figure size by 30%
plt.gcf().set_size_inches(10, 7)
plt.show()
```

```python
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(clf, X_train, y_train, cv = 10)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred, y_test))
```
